```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("Input/rubix.png")
img = cv2.cvtColor(img ,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# red
img[np.where(np.all((img == [0,255,255]),axis = 2))] = 255 , 0 , 0
# green
img[np.where(np.all((img == [255,0,255]),axis = 2))] = 0 , 255 , 0
# blue
img[np.where(np.all((img == [255,255,0]),axis = 2))] = 0 , 0 , 255

plt.imshow(img)
plt.imsave("Rubix.png" , img)
```
