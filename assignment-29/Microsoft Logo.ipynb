```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
import time
```

```python
img = np.ones((640,1360,3) , dtype = "uint8")*95
plt.imshow(img)
```

```python
#red
img[210:310,210:310] = (246, 83, 20)
#green
img[210:310,320:420] = (124, 187, 0)
#blue
img[320:420,210:310] = (0, 161, 241)
#yellow
img[320:420,320:420] = ( 255, 187, 0)
plt.imshow(img)
```

```python
font = ImageFont.truetype("arial.ttf", 160)
img_pil = Image.fromarray(img)
draw = ImageDraw.Draw(img_pil)
draw.text((460, 230),  "Microsoft", font = font, stroke_width=3 , fill = (255,255,255))
img = np.array(img_pil)
 
plt.imshow(img)
img_rgb = cv2.cvtColor(img ,cv2.COLOR_BGR2RGB)
cv2.imwrite("Microsoft Logo.png", img_rgb)
```
