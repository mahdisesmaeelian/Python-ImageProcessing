```python
import cv2
import matplotlib.pyplot as plt
```

```python
img =cv2.imread("input\SuperMan.jpg")
rgb_img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
background_img = cv2.imread("input\sky.jpg")
background_img = cv2.cvtColor(background_img , cv2.COLOR_BGR2RGB)
plt.imshow(background_img)
```

```python
img_hsv = cv2.cvtColor(img , cv2.COLOR_BGR2HSV)
H , S , V = cv2.split(img_hsv)
```

```python
rows , cols = H.shape
background = cv2.resize(background_img,(cols,rows))

for i in range (rows):
    for j in range (cols):
        if 37 < H[i,j] < 80 and S[i,j] > 30:
            rgb_img[i,j] = background[i,j]

plt.imshow(rgb_img)
```
