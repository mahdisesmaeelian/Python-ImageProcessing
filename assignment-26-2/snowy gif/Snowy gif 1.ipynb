```python
import random 
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imageio
import os
```

```python
image = cv2.imread('input/village-scaled.jpg')
```

```python
row , col , _= image.shape
snow_x_list = []
snow_y_list = []
snow_radius_list = []


for i in range(0,500):
    random_x = random.randint(-1000,row)
    random_y = random.randint(-1000,col)
    random_radius = random.randint(2,3)
    snow = cv2.circle(image,(random_y,random_x) ,random_radius, (255, 255, 255),-1)
    snow_x_list.append(random_x)
    snow_y_list.append(random_y)
    snow_radius_list.append(random_radius)

counter = 1
a = 100
for k in range(0,100):
    image = cv2.imread('input/village-scaled.jpg') 
    for j in range(0,500):
        snow = cv2.circle(image,(snow_y_list[j],snow_x_list[j]+counter) ,snow_radius_list[j], (255, 255, 255),-1)
    a += 1      
    counter += 10
    cv2.imwrite(f'output/img{a}.jpg',image)
        
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

files = os.listdir('output')
images=[]

for i in range (len(files)):
    image=imageio.imread('output/'+files[i])
    images.append(image)

imageio.mimsave('mygif.gif',images) 
```
