```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread( "Input/3.jpg")
img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap = "gray")
```

```python
img = cv2.GaussianBlur(img, (3, 3), 0)
lap = cv2.Laplacian(img ,cv2.CV_16S)
lap = np.uint8(np.absolute(lap))
plt.imshow(lap, cmap = "gray")
```

```python
_ , th1 = cv2.threshold(lap , 10 , 255 , cv2.THRESH_BINARY)
plt.imshow(th1, cmap = "gray")
```

```python
contours , _ = cv2.findContours(th1, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

for c in contours:
    if len(c) > 50:
        x, y, w, h = cv2.boundingRect(c)
        cv2.rectangle(img, (x,y) , (x+w,y+h), (127,127,127), 4)
        plt.imshow(img, cmap = "gray")
```
