```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread( "Input/0.jpg" , 0)
clr_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR);
plt.imshow(img, cmap = "gray")
```

```python
_ , th1 = cv2.threshold(img , 75 , 255 , cv2.THRESH_BINARY)
th1 = 255 - th1
plt.imshow(th1, cmap = "gray")
```

```python
contours , _ = cv2.findContours(th1, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
counter = 0
for c in contours:
    if len(c) > 100:
        x, y, w, h = cv2.boundingRect(c)
        cv2.rectangle(clr_img, (x,y) , (x+w,y+h), (0, 255, 0), 4)
        counter += 1

cv2.putText(clr_img, text= f'Dice number is : {counter}', org=(20, 50), fontFace=cv2.FONT_HERSHEY_TRIPLEX, fontScale=0.75, color=(0, 255, 0),thickness=2)
plt.imshow(clr_img, cmap = "gray")
```
