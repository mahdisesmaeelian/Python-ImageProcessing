```python
import cv2
import matplotlib.pyplot as plt
```

```python
org_img = cv2.imread( "Input/wolf.jpg" , cv2.IMREAD_GRAYSCALE)
_ , img = cv2.threshold(org_img , 120 , 255 , cv2.THRESH_BINARY)
```

```python
plt.imshow(img, cmap = "gray")
```

```python
# img = 255 - img
plt.imshow(img, cmap = "gray")
img_result = img.copy()
```

```python
contours , _ = cv2.findContours(img_result, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for c in contours:
    if len(c) > 200:
        x, y, w, h = cv2.boundingRect(c)
        cv2.rectangle(org_img, (x,y) , (x+w,y+h), (127,127,127), 4)
        plt.imshow(org_img, cmap = "gray")
```
